```python
## Feature scaling
def mean_normalization(X):
    mean = np.mean(X, axis=0)
    X_max = np.max(X, axis=0)
    X_min = np.min(X, axis=0)
    X_scaled = (X - mean) / (X_max - X_min)
    return X_scaled
```

```python
## Linear Regression multiple features
##----------------
# Read and plot training data
X_train, Y_train = read_data('./data/linear_regression_multi_feature_data.txt', feature_count=2)
plot_linear_data(X_train[:,0], Y_train)
plot_linear_data(X_train[:,1], Y_train)

X_scaled = mean_normalization(X_train)

# Plot the scaled data
plt.scatter(X_scaled[:,0], Y_train, marker = '.', c='orange')
plt.scatter(X_scaled[:,1], Y_train, marker = 'x', c='lime')
plt.title("Feature scaled data")
plt.show()
```

```python
# Gradient Descent
optimal_w, optimal_b, gd_log = gradient_descent(X_scaled, Y_train, w=np.zeros(X_scaled.shape[1]), b=0, alpha=0.01, iteration_count=1000, lambda_reg=0, log_verbosity=100, logistic=False)

# Plot the cost function against the iteration count
plt.plot(gd_log[:,0], gd_log[:,1])
plt.xlabel("Iteration count")
plt.title("Cost function (J)")
plt.show()

# Plot the training data predictions(Y_hat) side by side with the ground truth (Y_train)
Y_hat = prediction_function(X_scaled, optimal_w, optimal_b)
fig = plt.figure(figsize=(10, 8))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
ax1.set_title("X1_scaled , Y_train")
ax2.set_title("X1_scaled , Y_predicted")
ax3.set_title("X2_scaled , Y_train")
ax4.set_title("X2_scaled , Y_predicted")
ax1.scatter(X_scaled[:,0], Y_train, marker = '.', c='orange')
ax2.scatter(X_scaled[:,0], Y_hat, marker = '.', c='r')
ax3.scatter(X_scaled[:,1], Y_train, marker = 'x', c='lime')
ax4.scatter(X_scaled[:,1], Y_hat, marker = 'x', c='r')
plt.show()
```
